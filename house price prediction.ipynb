Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset

```python
home_rate_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(home_rate_dataset)
```

Loading the dataset into pandas dataframe

```python
home_rate_dataframe = pd.DataFrame(home_rate_dataset.data, columns = home_rate_dataset.feature_names)
```

```python
#print first 5 rows of the dataset
home_rate_dataframe.head()
```

```python
#Add the target (price) column to the dataframe
home_rate_dataframe['price'] = home_rate_dataset.target
```

```python
home_rate_dataframe.head()
```

```python
#Checking the number of rows and column
home_rate_dataframe.shape
```

```python
#check for missing value
home_rate_dataframe.isnull().sum()
```

```python
#Statistical measure of the dataset
home_rate_dataframe.describe()
```

```python
#splitting the data and the target
X = home_rate_dataframe.drop(['price'], axis = 1)
Y = home_rate_dataframe['price']
```

```python
print(X)
print(Y)
```

```python
#splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training
XGBoost Regressor

```python
#load the model
project = XGBRegressor()
```

```python
#training the model with X_train
project.fit(X_train, Y_train)
```

evaluation

Prediction of our training data

```python
#accuracy for prediction on training data
training_data_prediction = project.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R square error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
```

```python
print("R square error : ", score_1)
print("Mean absolute error : ", score_2)
```

Visualization of the acutal and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("actual prices")
plt.ylabel("predicted prices")
plt.title("actual vs predicted")
plt.show()
```

Prediction on Test data

```python
#accuracy for prediction on training data
test_data_prediction = project.predict(X_test)
```

```python
#R square error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

#mean absolute error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R square error : ", score_1)
print("Mean absoulte error : ", score_2)
```
